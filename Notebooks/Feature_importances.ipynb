# Libraries

```python
import matplotlib.pyplot as plt
from pathlib import Path
import sys
sys.path.append('../src')

from Models.model_wrappers import XGB_VotingRegressor, CatBst_Voting_Regressor

plt.style.use('seaborn-v0_8')
```

# Paths

```python
MODEL_PATH = Path().cwd().parents[0] / 'Models'
```

# Load Models

```python
model_architecture = 'CatBoost'
model_name = 'EDx1_Some_FE_5mins_3VR_CatBoost_optimized_CV_v2'
```

```python
if model_architecture == "XGBoost":

    model = XGB_VotingRegressor.load(MODEL_PATH/model_name)

elif model_architecture == "CatBoost":

    model = CatBst_Voting_Regressor.load(MODEL_PATH/model_name)
```

# Feature Importance

```python
n_feats = 20
```

```python
f_importances = model.get_feature_importances()

fig, ax = plt.subplots(2,2, figsize=(20,20))

f_importances_trucated = f_importances.iloc[0:n_feats].sort_values()
ax[0,0].barh(f_importances_trucated.index, f_importances_trucated.values)
ax[0,0].set_title(f'Top{n_feats}')

f_importances_trucated = f_importances.iloc[n_feats:n_feats*2].sort_values()
ax[0,1].barh(f_importances_trucated.index, f_importances_trucated.values)
ax[0,1].set_title(f'Top{n_feats} - Top{n_feats*2}')

f_importances_trucated = f_importances.iloc[n_feats*2:n_feats*3].sort_values()
ax[1,0].barh(f_importances_trucated.index, f_importances_trucated.values)
ax[1,0].set_title(f'Top{n_feats*2} - Top{n_feats*3}')

f_importances_trucated = f_importances.iloc[n_feats*3:n_feats*4].sort_values()
ax[1,1].barh(f_importances_trucated.index, f_importances_trucated.values)
ax[1,1].set_title(f'Top{n_feats*3} - Top{n_feats*4}')
```
